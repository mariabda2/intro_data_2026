### <span style="color:#2F749F;"><strong>üìö Definici√≥n de Ciencia de Datos</strong></span>

La ciencia de datos es un campo <span style="color:#01A020;"><strong>interdisciplinario</strong></span> que combina m√©todos cient√≠ficos, algoritmos, procesos y sistemas para <span style="color:#F7AA46;"><strong>extraer conocimiento</strong></span>  y <span style="color:#337AB7;"><strong>generar valor</strong></span>  a partir de datos estructurados y no estructurados. Integra principios de estad√≠stica, inform√°tica, aprendizaje autom√°tico y an√°lisis de datos, con el objetivo de **construir modelos predictivos, descubrir patrones y respaldar** la toma de decisiones en diversos entornos.

Seg√∫n Dhar (2013), la ciencia de datos no solo implica el manejo de grandes vol√∫menes de datos (big data), sino tambi√©n la **transformaci√≥n de estos** en informaci√≥n mediante el uso de t√©cnicas anal√≠ticas avanzadas. En esta l√≠nea, Provost y Fawcett (2013) enfatizan que la ciencia de datos se centra en el desarrollo de conocimientos relevantes para **resolver problemas reales**, m√°s all√° del simple an√°lisis exploratorio. Asimismo, Donoho (2017) destaca que la ciencia de datos **representa una evoluci√≥n natural** de la estad√≠stica en la era del c√≥mputo intensivo, donde el procesamiento masivo y la visualizaci√≥n juegan roles fundamentales en la comprensi√≥n de fen√≥menos complejos.

> **üìú Origen y evoluci√≥n de la ciencia de datos.** El t√©rmino "ciencia de datos" se atribuye originalmente al estad√≠stico Peter Naur, quien lo utiliz√≥ en 1960 como una alternativa al concepto de "procesamiento de datos". M√°s tarde, en 1974, el mismo Naur public√≥ el libro Concise Survey of Computer Methods, donde formaliz√≥ su visi√≥n. Sin embargo, fue en la d√©cada de 1990 cuando el t√©rmino comenz√≥ a tomar fuerza, especialmente con el trabajo de William S. Cleveland, quien propuso formalmente establecer la ciencia de datos como una disciplina independiente, integrando estad√≠sticas con computaci√≥n y almacenamiento masivo de datos. Desde entonces, la disciplina ha evolucionado con el auge del big data, el aprendizaje autom√°tico y la inteligencia artificial, posicion√°ndose como un eje estrat√©gico en m√∫ltiples sectores como la salud, finanzas, educaci√≥n, y tecnolog√≠a.

```python
# Importaci√≥n de librer√≠as
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

### <span style="color:#2F749F;"><strong> ü©∫ Caso 1: Predicci√≥n de enfermedades cr√≥nicas</strong></span>

- **Contexto:**  Un hospital quiere anticipar qu√© pacientes tienen mayor riesgo de desarrollar **diabetes tipo 2** en los pr√≥ximos 5 a√±os.
- **Datos disponibles:**  Edad, IMC, Presi√≥n arterial, Historial familiar, Nivel de glucosa, Actividad f√≠sica

- **Aplicaci√≥n:**  Se entrena un modelo de clasificaci√≥n (por ejemplo, `RandomForestClassifier`) para predecir **riesgo alto vs bajo**.   El modelo permite **priorizar intervenciones preventivas** y campa√±as de salud p√∫blica.

```python
# Simulaci√≥n de datos cl√≠nicos
# Simulaci√≥n de datos cl√≠nicos
np.random.seed(42)
n = 500
df_diabetes = pd.DataFrame({
    'edad': np.random.randint(20, 70, n),
    'imc': np.random.normal(28, 5, n),
    'presion': np.random.normal(120, 15, n),
    'glucosa': np.random.normal(100, 20, n),
    'actividad': np.random.randint(0, 5, n),
    'riesgo': np.random.choice([0, 1], size=n, p=[0.7, 0.3])
})

# Separaci√≥n de variables
X = df_diabetes.drop('riesgo', axis=1)
y = df_diabetes['riesgo']

# Divisi√≥n en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Modelo Random Forest
modelo_rf = RandomForestClassifier()
modelo_rf.fit(X_train, y_train)

# Predicci√≥n y evaluaci√≥n
y_pred = modelo_rf.predict(X_test)
print("Reporte de clasificaci√≥n:")
print(classification_report(y_test, y_pred))

# Visualizaci√≥n 
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df_diabetes[df_diabetes['riesgo'] == 0]['glucosa'], ax=axes[0], color='skyblue', kde=True)
axes[0].set_title("Glucosa - Riesgo Bajo")
sns.histplot(df_diabetes[df_diabetes['riesgo'] == 1]['glucosa'], ax=axes[1], color='salmon', kde=True)
axes[1].set_title("Glucosa - Riesgo Alto")
plt.suptitle("Distribuci√≥n de glucosa seg√∫n riesgo de diabetes")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_diabetes,
    x='imc',
    y='glucosa',
    hue='riesgo',
    palette={0: 'skyblue', 1: 'salmon'}
)
plt.title("Relaci√≥n entre IMC y Glucosa seg√∫n riesgo")
plt.xlabel("IMC")
plt.ylabel("Glucosa")
plt.legend(title="Riesgo", loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()

# Presi√≥n arterial por riesgo
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df_diabetes,
    x='riesgo',
    y='presion',
    hue='riesgo',  # Se asigna hue expl√≠citamente
    palette={0: 'skyblue', 1: 'salmon'},
    dodge=False  # Para evitar duplicaci√≥n de cajas
)
plt.title("Distribuci√≥n de presi√≥n arterial por riesgo")
plt.xlabel("Riesgo de diabetes")
plt.ylabel("Presi√≥n arterial")
plt.legend(title="Riesgo", loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlaci√≥n
plt.figure(figsize=(8, 6))
corr = df_diabetes.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de calor de correlaciones cl√≠nicas")
plt.tight_layout()
plt.show()
```

- ¬øQu√© pasa si el modelo tiene **sesgo por grupo √©tnico o g√©nero**?  
- ¬øC√≥mo se comunica el riesgo sin **generar alarma innecesaria**?

### <span style="color:#2F749F;"><strong>üí≥ Caso 2: Detecci√≥n de Fraude Bancario</strong></span>

- **Contexto:** Una fintech quiere identificar transacciones sospechosas en tiempo real para prevenir fraudes.
- **Datos disponibles:** Monto, hora, ubicaci√≥n, tipo de comercio, historial del usuario.
- **Aplicaci√≥n:** Se usa un modelo de detecci√≥n de anomal√≠as (por ejemplo, IsolationForest) para marcar transacciones inusuales. Las alertas se env√≠an al equipo de seguridad para revisi√≥n.

```python
# Simulaci√≥n de datos
n = 500
np.random.seed(42)
df_fraude = pd.DataFrame({
    'monto': np.random.exponential(scale=100, size=n),
    'hora': np.random.randint(0, 24, n),
    'ubicacion': np.random.randint(1, 10, n),
    'comercio': np.random.randint(1, 20, n)
})

# Detecci√≥n de anomal√≠as
modelo_iso = IsolationForest(contamination=0.05)
df_fraude['anomal√≠a'] = modelo_iso.fit_predict(df_fraude)


# Visualizaci√≥n
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_fraude,
    x='monto',
    y='hora',
    hue='anomal√≠a',
    palette={1: 'gray', -1: 'red'}
)
plt.title("Transacciones normales vs sospechosas")
plt.xlabel("Monto")
plt.ylabel("Hora del d√≠a")
plt.legend(title="Tipo", loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()

# Histograma
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
variables = ['monto', 'hora', 'ubicacion', 'comercio']
colors = ['steelblue', 'salmon', 'mediumseagreen', 'orchid']

for ax, var, color in zip(axs.flat, variables, colors):
    sns.histplot(df_fraude[var], bins=30, kde=True, ax=ax, color=color)
    ax.set_title(f'Distribuci√≥n de {var}')
    ax.grid(True)

plt.tight_layout()
plt.show()

# Detecci√≥n de at√≠picos
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_fraude[['monto', 'hora', 'ubicacion', 'comercio']], palette='Set2')
plt.title("Boxplots de variables transaccionales")
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlaci√≥n
plt.figure(figsize=(8, 6))
corr = df_fraude.drop(columns='anomal√≠a').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de calor de correlaciones")
plt.tight_layout()
plt.show()
```

#### <span style="color:#2F749F;"><strong>Discusi√≥n √©tica</strong></span>
- ¬øQu√© pasa si se bloquea una compra leg√≠tima?
- ¬øC√≥mo se balancea precisi√≥n vs rapidez?

### <span style="color:#2F749F;"><strong>üé¨ Caso 3: Recomendaci√≥n de contenido en una plataforma educativa</strong></span>

- **Contexto:** Una plataforma de cursos quiere sugerir nuevos contenidos relevantes a cada estudiante.
- **Datos disponibles:** Cursos vistos, tiempo de interacci√≥n, calificaciones, temas favoritos.
- **Aplicaci√≥n:** Se aplica un sistema de recomendaci√≥n basado en similitud de usuarios (Collaborative Filtering) o de contenidos (Content-Based Filtering). El sistema personaliza la experiencia y mejora la retenci√≥n.

```python
# Simulaci√≥n de preferencias
np.random.seed(42)
usuarios = [f'User_{i}' for i in range(1, 11)]
cursos = [f'Course_{j}' for j in range(1, 6)]
preferencias = np.random.randint(1, 6, size=(10, 5))  # valores de 1 a 5

df = pd.DataFrame(preferencias, index=usuarios, columns=cursos)

# Popularidad de cursos
plt.figure(figsize=(8, 5))
df.mean().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Popularidad promedio de cursos')
plt.ylabel('Puntuaci√≥n promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Preferencias individuales 
plt.figure(figsize=(6, 6))
df.loc['User_1'].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribuci√≥n de preferencias de User_1')
plt.ylabel('')
plt.tight_layout()
plt.show()

# Similitud entre usuarios
sim_matrix = cosine_similarity(df)
sim_df = pd.DataFrame(sim_matrix, index=usuarios, columns=usuarios)
sim_long = sim_df.stack().reset_index()
sim_long.columns = ['User_A', 'User_B', 'Similarity']
sim_long = sim_long[sim_long['User_A'] != sim_long['User_B']]  # excluir diagonal

plt.figure(figsize=(8, 6))
sns.scatterplot(data=sim_long, x='User_A', y='User_B', hue='Similarity', palette='viridis', size='Similarity', sizes=(50, 200))
plt.title('Similitud entre usuarios basada en preferencias')
plt.xticks(rotation=45)
plt.legend(title="Similitud", loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

#### <span style="color:#2F749F;"><strong>Discusi√≥n √©tica</strong></span>
- ¬øQu√© pasa si el sistema refuerza sesgos o limita la diversidad de temas?
- ¬øC√≥mo se evita la ‚Äúburbuja de contenido‚Äù?

## <span style="color:#2F749F;"><strong>Falsos Positivos y Falsos Negativos</strong></span>
---

### <span style="color:#2F749F;"><strong>Definiciones r√°pidas</strong></span>
- **<span style="color:black;">Falso Positivo (FP)</span>**: El detector predice *positivo* (hay fuego) cuando en realidad es *negativo* (no hay fuego).  
- **<span style="color:black;">Falso Negativo (FN)</span>**: El detector predice *negativo* (no hay fuego) cuando en realidad es *positivo* (s√≠ hay fuego).  


### <span style="color:#2F749F;"><strong>Truco para recordarlo</strong></span>
- **Positivo / Negativo** ‚Üí lo que el detector *predice*.  
- **Falso** ‚Üí el detector *se equivoc√≥*.  
- **Falso positivo** ‚Üí dijo ‚Äúpositivo‚Äù pero estaba mal.  
- **Falso negativo** ‚Üí dijo ‚Äúnegativo‚Äù pero estaba mal.
