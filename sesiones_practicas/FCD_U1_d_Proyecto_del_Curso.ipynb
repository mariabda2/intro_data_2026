## <span style="color:Blue;"><strong>Preparaci√≥n de la fuente de datos y librer√≠as</strong></span>

```python
# Librer√≠as
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo general
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Datos
# URL directa al archivo ZIP que contiene los datos
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip"

# Descarga y extracci√≥n
import zipfile, io, requests

r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall("data/student_data")

# Crear dataset
df = pd.read_csv("data/student_data/student-mat.csv", sep=";")
```

## <span style="color:Blue;"><strong>4. Visualizaciones exploratorias b√°sicas</strong></span>

<div style="background-color:#FCE4EC; padding:10px; border-radius:8px;">
<b>- Gr√°ficos de dispersi√≥n para relaciones entre variables.</b><br>
<b>- Breve interpretaci√≥n de lo que muestran las gr√°ficas.</b><br>
<b>- Histogramas o gr√°ficos de barras para variables num√©ricas y categ√≥ricas.</b><br>
</div>


```python
# Histograma: distribuci√≥n de edad
sns.histplot(df['age'], bins=6, kde=True, color='skyblue')
plt.title("Distribuci√≥n de edad de los estudiantes")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Gr√°fico de barras: tiempo de estudio
# Mapeo de categor√≠as
studytime_labels = {
    1: "<2 h/sem",
    2: "2‚Äì5 h/sem",
    3: "5‚Äì10 h/sem",
    4: ">10 h/sem"
}

# Columna con etiquetas
df['studytime_label'] = df['studytime'].map(studytime_labels)

# Gr√°fico 
sns.countplot(x='studytime_label', data=df, hue='studytime_label', palette='viridis',legend=False)
plt.title("Tiempo de estudio semanal")
plt.xlabel("Categor√≠a de tiempo de estudio")
plt.ylabel("N√∫mero de estudiantes")
plt.show()
```

```python
# Gr√°fico de barras: g√©nero
sns.countplot(x='sex', data=df,hue='sex', palette='pastel',legend=False)
plt.title("Distribuci√≥n por g√©nero")
plt.xlabel("G√©nero")
plt.ylabel("N√∫mero de estudiantes")
plt.show()
```

```python
# Gr√°fico de dispersi√≥n: calificaci√≥n final vs ausencias
sns.scatterplot(x='absences', y='G3', hue='sex', data=df)
plt.title("Relaci√≥n entre ausencias y calificaci√≥n final (G3)")
plt.xlabel("N√∫mero de ausencias")
plt.ylabel("Calificaci√≥n final (G3)")
plt.show()
```

```python
# Gr√°fico de dispersi√≥n: tiempo de estudio vs calificaci√≥n final
sns.boxplot(x='studytime_label', y='G3',hue='studytime_label', data=df, palette='coolwarm')
plt.title("Calificaci√≥n final seg√∫n tiempo de estudio")
plt.xlabel("Tiempo de estudio")
plt.ylabel("Calificaci√≥n final (G3)")
plt.show()
```

```python
# Gr√°fico de dispersi√≥n con l√≠neas de tendencia
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# G1 vs G2
sns.regplot(x='G1', y='G2', data=df, ax=axes[0],
            scatter_kws={'s': 50, 'alpha': 0.7},
            line_kws={'color': 'red'})
axes[0].set_title('Relaci√≥n entre G1 y G2', fontsize=14)
axes[0].set_xlabel('Nota Parcial 1 (G1)', fontsize=12)
axes[0].set_ylabel('Nota Parcial 2 (G2)', fontsize=12)

# G2 vs G3
sns.regplot(x='G2', y='G3', data=df, ax=axes[1],
            scatter_kws={'s': 50, 'alpha': 0.7},
            line_kws={'color': 'blue'})
axes[1].set_title('Relaci√≥n entre G2 y Nota Final (G3)', fontsize=14)
axes[1].set_xlabel('Nota Parcial 2 (G2)', fontsize=12)
axes[1].set_ylabel('Nota Final (G3)', fontsize=12)

plt.tight_layout()
plt.show()
```
