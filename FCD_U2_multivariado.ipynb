# <span style="color:#2F749F;"><strong>3. An√°lisis Multivariado</strong></span>

El an√°lisis multivariado se refiere al estudio simult√°neo de **tres o m√°s variables**, con el prop√≥sito de **identificar patrones, dependencias complejas y estructuras ocultas** en los datos. Este tipo de an√°lisis es fundamental en contextos donde las variables no act√∫an de forma independiente, sino que se encuentran relacionadas de manera conjunta [(Hair, 2019)](https://eli.johogo.com/Class/CCU/SEM/_Multivariate%20Data%20Analysis_Hair.pdf).

> Este an√°lisis permite:
> 
>- Detectar **relaciones multidimensionales** entre variables.  
>- Reducir la complejidad de los datos a trav√©s de t√©cnicas de **reducci√≥n de dimensionalidad**.  
>- Construir modelos que expliquen fen√≥menos complejos.  
>- Aplicar m√©todos predictivos m√°s robustos (Rencher & Christensen, 2012).  

Entre las principales herramientas del an√°lisis multivariado se encuentran:  
1. **An√°lisis de Componentes Principales (ACP / PCA)**  
2. **Regresi√≥n M√∫ltiple**  
3. **An√°lisis de Cl√∫ster (Cluster Analysis)**  
4. **An√°lisis Factorial**


## <span style="color:black;"><strong>3.1. An√°lisis de Componentes Principales (PCA)</strong></span> 

Este an√°lisis busca transformar un conjunto de variables correlacionadas en un n√∫mero menor de **componentes no correlacionados**, maximizando la varianza explicada [(Jolliffe & Cadima, 2016)](https://doi.org/10.1098/rsta.2015.0202).  

$$
\mathbf{Z} = \mathbf{XW}
$$  

donde:  
- $\mathbf{X}$ es la matriz de datos estandarizados ($n \times p$).  
- $\mathbf{W}$ es la matriz de **autovectores** de la matriz de covarianzas.  
- $\mathbf{Z}$ es la matriz de componentes principales.  
 
### <span style="color:#2F749F;"><strong>Fundamentos matem√°ticos</strong></span>  

El **An√°lisis de Componentes Principales (PCA)** se fundamenta en la **descomposici√≥n espectral** de la matriz de covarianzas o de correlaciones [(Abdi & Williams, 2010)](https://doi.org/10.1002/wics.101). El objetivo es encontrar una transformaci√≥n ortogonal que proyecte los datos en un nuevo espacio donde:  

1. Los componentes principales (PCs) son **lineales** de las variables originales.  
2. Cada componente maximiza la **varianza** de los datos en esa direcci√≥n.  
3. Los componentes son **ortogonales** entre s√≠, lo que elimina la multicolinealidad.  

Matem√°ticamente:  

1. Se calcula la **matriz de covarianzas**:  

$$
\mathbf{S} = \frac{1}{n-1} \mathbf{X}^\top \mathbf{X}
$$  

2. Se obtienen **autovalores** $(\lambda_1, \lambda_2, \dots, \lambda_p)$ y **autovectores** $(\mathbf{w}_1, \mathbf{w}_2, \dots, \mathbf{w}_p)$ de $\mathbf{S}$.  
3. El primer componente principal corresponde al autovector con el mayor autovalor, y as√≠ sucesivamente [(Shlens, 2014)](https://doi.org/10.48550/arXiv.1404.1100).  



### <span style="color:#2F749F;"><strong>Interpretaci√≥n de los componentes</strong></span>  

- **Autovalores ($\lambda_i$):** representan la cantidad de varianza explicada por cada componente.  
- **Autovectores ($\mathbf{w}_i$):** son las direcciones de proyecci√≥n (pesos asignados a cada variable).  
- **Proporci√≥n de varianza explicada (PVE):**  

$$
\text{PVE}_i = \frac{\lambda_i}{\sum_{j=1}^{p} \lambda_j}
$$  

La PVE se utiliza para decidir cu√°ntos componentes retener. Una pr√°ctica com√∫n es elegir el n√∫mero de componentes que explican al menos el **70‚Äì90% de la varianza total** [(Jolliffe & Cadima, 2016)](https://doi.org/10.1098/rsta.2015.0202).

<div style="background-color:#fff9db; padding:10px; border-radius:5px;">
  <b>Cargar datos reales (Iris), estandarizaci√≥n y PCA</b><br>
</div>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Cargar datos Iris
iris = load_iris()

# Crear DataFrame 
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Mostrar primeras filas
print(df.head())

# Estandarizar variables num√©ricas
scaler = StandardScaler()
X_std = scaler.fit_transform(df[iris.feature_names])

# PCA
pca = PCA()
X_pca = pca.fit_transform(X_std)

sns.pairplot(df, hue='species', palette='Set2', diag_kind='hist', corner=False)
plt.suptitle('Matriz de dispersi√≥n: relaciones entre variables del Iris', fontsize=16, y=1.02)
plt.show()
```

<div style="background-color:#fff9db; padding:20px; border-radius:10px;">
<b>Paso 1:</b> Los componentes principales son combinaciones lineales de las variables originales
</div>

```python
# Matriz de componentes
componentes = pd.DataFrame(pca.components_.T, 
                           columns=[f'PC{i+1}' for i in range(len(iris.feature_names))], 
                           index=iris.feature_names)

# üìê Visualizaci√≥n de autovectores
plt.figure(figsize=(8, 5))
sns.heatmap(componentes, annot=True, cmap='coolwarm', center=0)
plt.title('Autovectores')
plt.ylabel('Variables originales')
plt.xlabel('Componentes principales')
plt.tight_layout()
plt.show()

# Ecuaciones de los componentes principales con nombres reales
print("Ecuaciones de los componentes principales:\n")
for i, comp in enumerate(componentes.columns):
    pesos = pca.components_[i]
    ecuacion = " + ".join([f"{round(pesos[j], 3)}¬∑'{iris.feature_names[j]}'" for j in range(len(pesos))])
    print(f"{comp} = {ecuacion}")
```

<div style="background-color:#fff9db; padding:20px; border-radius:10px;">
<b>Paso 2:</b> Cada componente maximiza la varianza en su direcci√≥n
</div>

```python
# Varianza explicada
varianza = pca.explained_variance_
PVE = pca.explained_variance_ratio_
PVE_acum = np.cumsum(PVE)
componentes = [f'PC{i+1}' for i in range(len(PVE))]

# DataFrame para visualizaci√≥n
df_varianza = pd.DataFrame({
    'Componente': componentes,
    'Autovalor': varianza,
    'PVE': PVE,
    'PVE acumulada': PVE_acum
})

# Autovalores
plt.figure(figsize=(8, 5))
sns.barplot(x='Componente', y='Autovalor', data=df_varianza, color='lightsteelblue')
plt.title('Autovalores')
plt.ylabel('Autovalor (Œª)')
plt.xlabel('Componentes principales')
plt.grid(True)
plt.tight_layout()
plt.show()


# Proporci√≥n de varianza explicada
plt.figure(figsize=(8, 5))
sns.barplot(x='Componente', y='PVE', data=df_varianza, color='skyblue', label='PVE individual')
sns.lineplot(x='Componente', y='PVE acumulada', data=df_varianza, marker='o', color='darkblue', label='PVE acumulada')
plt.axhline(0.7, color='red', linestyle='--', label='Umbral 70%')
plt.axhline(0.9, color='green', linestyle='--', label='Umbral 90%')
plt.title('Proporci√≥n de varianza explicada por componente')
plt.ylabel('Proporci√≥n de varianza')
plt.legend()
plt.tight_layout()
plt.show()
```

<div style="background-color:#fff9db; padding:20px; border-radius:10px;">
<b>Paso 3:</b> Los componentes son ortogonales entre s√≠
</div>

```python
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(len(PVE))])
correlaciones = df_pca.corr()

# Visualizaci√≥n
plt.figure(figsize=(6, 5))
sns.heatmap(correlaciones, annot=True, cmap='vlag', center=0)
plt.title('Correlaci√≥n entre componentes principales')
plt.tight_layout()
plt.show()

df_pca['species'] = df['species']

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='species', palette='Set2')
plt.title('Proyecci√≥n de los datos en los dos primeros componentes')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.tight_layout()
plt.show()
```



### <span style="color:#2F749F;"><strong>Ventajas y limitaciones</strong></span>  

**Ventajas:**  
- Reduce la dimensionalidad sin p√©rdida excesiva de informaci√≥n.  
- Elimina la multicolinealidad entre variables.  
- Facilita la visualizaci√≥n de datos en 2D o 3D [(Lever, Krzywinski, & Altman, 2017)](https://doi.org/10.1038/nmeth.4346).  

**Limitaciones:**  
- Los componentes son combinaciones lineales, lo que dificulta la interpretaci√≥n directa.  
- PCA es sensible a la **escala de las variables** (es recomendable estandarizar antes de aplicar).  
- Puede no ser adecuado si los datos presentan relaciones no lineales [(Abdi & Williams, 2010)](https://doi.org/10.1002/wics.101).  


### <span style="color:#2F749F;"><strong>Aplicaciones pr√°cticas</strong></span>  

- **Compresi√≥n de datos:** reducci√≥n de caracter√≠sticas en im√°genes (e.g., reconocimiento facial) [(Paricio et al., 2025)](https://doi.org/10.1016/j.asoc.2025.112877).  
- **Exploraci√≥n de datos:** identificaci√≥n de patrones y cl√∫steres ocultos [(MeilƒÉ & Zhang, 2024)](https://doi.org/10.7717/peerj-cs.3025).  
- **Preprocesamiento en Machine Learning:** como paso previo a algoritmos de clasificaci√≥n o regresi√≥n [(Ramos-P√©rez et al., 2024)](https://doi.org/10.3390/info15040223).  
- **Bioinform√°tica:** an√°lisis de expresi√≥n g√©nica y reducci√≥n de dimensiones en datos √≥micos [(Jolliffe & Cadima, 2016)](https://doi.org/10.1098/rsta.2015.0202).  


## <span style="color:black;"><strong>3.2. Regresi√≥n Lineal M√∫ltiple</strong></span>

La **regresi√≥n lineal m√∫ltiple** es una t√©cnica estad√≠stica utilizada para modelar la relaci√≥n entre una variable dependiente continua $y$ y un conjunto de variables independientes o predictores $x_1, x_2, \dots, x_p$ [(James, Witten, Hastie, & Tibshirani, 2021)](https://doi.org/10.1007/978-1-0716-1418-1).  

El modelo general se expresa como:  

$$
y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_p x_{ip} + \epsilon_i
$$  

donde:  
- $y_i$ es la variable dependiente para la observaci√≥n $i$.  
- $x_{ij}$ es el valor del predictor $j$ en la observaci√≥n $i$.  
- $\beta_j$ son los coeficientes de regresi√≥n que cuantifican el efecto de cada predictor.  
- $\epsilon_i$ es el t√©rmino de error, que representa la variaci√≥n no explicada por el modelo.  

### <span style="color:#2F749F;"><strong>Fundamentos matem√°ticos</strong></span>  

En notaci√≥n matricial, el modelo se representa como:  

$$
\mathbf{y} = \mathbf{X\beta} + \mathbf{\epsilon}
$$  

donde:  
- $\mathbf{y}$ es un vector $n \times 1$ de respuestas.  
- $\mathbf{X}$ es la matriz de dise√±o $n \times (p+1)$ (incluyendo la columna de unos para el intercepto).  
- $\mathbf{\beta}$ es un vector $(p+1) \times 1$ de par√°metros desconocidos.  
- $\mathbf{\epsilon}$ es el vector de errores aleatorios.  

El estimador de **M√≠nimos Cuadrados Ordinarios (OLS)** se obtiene minimizando la suma de errores cuadrados:  

$$
\hat{\beta} = (\mathbf{X}^\top \mathbf{X})^{-1} \mathbf{X}^\top \mathbf{y}
$$  

Este estimador es insesgado bajo los supuestos cl√°sicos de regresi√≥n [(Kutner, Nachtsheim, Neter, & Li, 2005)](https://www.mheducation.com/highered/product/applied-linear-statistical-models-kutner-nachtsheim/M9780073108742.html).  

### <span style="color:#2F749F;"><strong>Interpretaci√≥n de los coeficientes</strong></span>  

- $\beta_0$: valor esperado de $y$ cuando todas las variables predictoras son cero.  
- $\beta_j$: cambio esperado en $y$ por un incremento unitario en $x_j$, manteniendo constantes las dem√°s variables.  
- $R^2$: proporci√≥n de varianza de $y$ explicada por el modelo.  
- $R^2$ ajustado: corrige $R^2$ por el n√∫mero de predictores incluidos en el modelo.  

### <span style="color:#2F749F;"><strong>Supuestos del modelo</strong></span>  

Para que los resultados de RLM sean v√°lidos, deben cumplirse los siguientes supuestos [(Montgomery, Peck, & Vining, 2021)](https://doi.org/10.1002/9781119664036):  


1. **Linealidad:** relaci√≥n lineal entre predictores y la variable respuesta [(Kutner, Nachtsheim, & Neter, 2005)](https://doi.org/10.1002/0470052139). T√©cnicas de evaluaci√≥n:  
     - **Gr√°ficos de dispersi√≥n:** permiten visualizar si la relaci√≥n entre cada predictor y la variable dependiente sigue un patr√≥n aproximadamente recto.  
     - **Gr√°fico de residuos vs. valores ajustados:** si la relaci√≥n es lineal, los residuos deben distribuirse de forma aleatoria alrededor de cero, sin patrones curvil√≠neos.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

np.random.seed(42)
n = 200
X = np.random.normal(50, 10, n)

# Caso 1: Relaci√≥n lineal
Y_lineal = 3 + 0.8 * X + np.random.normal(0, 5, n)

# Caso 2: Relaci√≥n no lineal (cuadr√°tica)
Y_nolineal = 3 + 0.8 * X - 0.05 * X**2 + np.random.normal(0, 5, n)

df = pd.DataFrame({"X": X, "Y_lineal": Y_lineal, "Y_nolineal": Y_nolineal})

# Modelo lineal
X_lin = sm.add_constant(df["X"])
model_lin = sm.OLS(df["Y_lineal"], X_lin).fit()

# Modelo no lineal (ajustado como si fuera lineal)
model_nolin = sm.OLS(df["Y_nolineal"], X_lin).fit()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Caso lineal
# Dispersi√≥n
sns.scatterplot(x=df["X"], y=df["Y_lineal"], ax=axes[0, 0], color='steelblue')
axes[0, 0].plot(df["X"], model_lin.fittedvalues, color='red', label='Ajuste lineal')
axes[0, 0].set_title("Relaci√≥n lineal: Y vs X")
axes[0, 0].set_xlabel("Variable predictora X")
axes[0, 0].set_ylabel("Variable respuesta Y")
axes[0, 0].legend()

# Residuos
sns.scatterplot(x=model_lin.fittedvalues, y=model_lin.resid, ax=axes[1, 0], color='steelblue')
axes[1, 0].axhline(0, color='gray', linestyle='--')
axes[1, 0].set_title("Residuos vs valores ajustados (lineal)")
axes[1, 0].set_xlabel("Valores ajustados por el modelo")
axes[1, 0].set_ylabel("Residuos (Y - ≈∂)")

# Caso no lineal
# Dispersi√≥n
sns.scatterplot(x=df["X"], y=df["Y_nolineal"], ax=axes[0, 1], color='darkorange')
axes[0, 1].plot(df["X"], model_nolin.fittedvalues, color='red', label='Ajuste lineal')
axes[0, 1].set_title("Relaci√≥n no lineal: Y vs X")
axes[0, 1].set_xlabel("Variable predictora X")
axes[0, 1].set_ylabel("Variable respuesta Y")
axes[0, 1].legend()

# Residuos
sns.scatterplot(x=model_nolin.fittedvalues, y=model_nolin.resid, ax=axes[1, 1], color='darkorange')
axes[1, 1].axhline(0, color='gray', linestyle='--')
axes[1, 1].set_title("Residuos vs valores ajustados (no lineal)")
axes[1, 1].set_xlabel("Valores ajustados por el modelo")
axes[1, 1].set_ylabel("Residuos (Y - ≈∂)")

plt.tight_layout()
plt.show()
```

<div style="background-color:#f2f2f2; padding:10px; border-radius:5px;">
  <b>Interpretaci√≥n: supuesto linealidad</b><br>
  <ul>
    <li>Si los puntos en los gr√°ficos de dispersi√≥n siguen una tendencia recta, el supuesto se cumple.</li>
    <li>En el gr√°fico de residuos, si los puntos se distribuyen aleatoriamente alrededor de la l√≠nea cero, sin formar curvas ni patrones, la relaci√≥n es lineal.</li>
    <li>Si se observan formas en U, ondas o agrupamientos sistem√°ticos, puede ser necesario transformar variables o considerar modelos no lineales.</li>
  </ul>
</div>


2. **Independencia:** los errores deben ser independientes entre s√≠ (sin correlaci√≥n temporal o espacial) [(Durbin & Watson, 1950)](https://doi.org/10.2307/2332391). T√©cnicas de evaluaci√≥n:  
     - **Prueba de Durbin-Watson:** eval√∫a autocorrelaci√≥n en los residuos (valor cercano a 2 indica independencia; valores <1 o >3 sugieren autocorrelaci√≥n).  
     - **ACF/PACF en series temporales:** examinan correlaciones de los residuos en diferentes rezagos.   

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Simulaci√≥n de datos con posible autocorrelaci√≥n
np.random.seed(42)
n = 100
X = np.linspace(0, 10, n)
error_autocorrelado = np.zeros(n)
for t in range(1, n):
    error_autocorrelado[t] = 0.6 * error_autocorrelado[t-1] + np.random.normal(0, 1)

Y = 2 + 1.5 * X + error_autocorrelado
df = pd.DataFrame({"X": X, "Y": Y})

# Ajuste del modelo
X_const = sm.add_constant(df["X"])
model = sm.OLS(df["Y"], X_const).fit()
resid = model.resid

# Prueba de Durbin-Watson
dw = durbin_watson(resid)
print(f"Durbin-Watson: {dw:.3f} ‚Üí ¬øCercano a 2?")

# gr√°ficos de ACF (Autocorrelation Function) y PACF (Partial Autocorrelation Function)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(resid, ax=axes[0], lags=20)
axes[0].set_title("ACF de residuos")
plot_pacf(resid, ax=axes[1], lags=20)
axes[1].set_title("PACF de residuos")
plt.tight_layout()
plt.show()
```

<div style="background-color:#f2f2f2; padding:10px; border-radius:5px;">
  <b>Interpretaci√≥n: Supuesto de independencia</b><br>
  <ul>
    <li>La prueba de Durbin-Watson eval√∫a si los errores est√°n autocorrelacionados. Un valor cercano a 2 indica independencia; valores menores a 1 o mayores a 3 sugieren autocorrelaci√≥n.</li>
    <li>Los gr√°ficos ACF y PACF muestran correlaciones en distintos rezagos. Si hay picos significativos en los primeros rezagos, se confirma la presencia de autocorrelaci√≥n.</li>
    <li>La violaci√≥n de este supuesto puede afectar la validez de los intervalos de confianza y las pruebas de significancia del modelo.</li>
  </ul>
</div>



3. **Homoscedasticidad:** la varianza de los errores debe ser constante a lo largo de los valores predichos [(Breusch & Pagan, 1979)](https://doi.org/10.2307/2347650). T√©cnicas de evaluaci√≥n:
     - **Gr√°fico de residuos vs. valores ajustados:** si aparece un patr√≥n en forma de cono o embudo, indica heterocedasticidad.  
     - **Prueba de Breusch‚ÄìPagan:** contrasta la varianza de los residuos respecto a las variables explicativas (p-valor bajo ‚Üí heterocedasticidad).  
     - **Prueba de White:** similar a Breusch‚ÄìPagan, pero m√°s general al no depender de la forma funcional de la heterocedasticidad.  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan


np.random.seed(42)
n = 200
X = np.random.normal(50, 10, n)

# Caso 1: Homoscedasticidad
error_homo = np.random.normal(0, 5, n)
Y_homo = 2 + 0.5 * X + error_homo

# Caso 2: Heterocedasticidad
error_hetero = np.random.normal(0, X * 0.1)  # varianza proporcional a X
Y_hetero = 2 + 0.5 * X + error_hetero

df = pd.DataFrame({"X": X, "Y_homo": Y_homo, "Y_hetero": Y_hetero})
X_const = sm.add_constant(df["X"])

model_homo = sm.OLS(df["Y_homo"], X_const).fit()
model_hetero = sm.OLS(df["Y_hetero"], X_const).fit()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Homoscedasticidad
sns.scatterplot(x=model_homo.fittedvalues, y=model_homo.resid, ax=axes[0], color='steelblue')
axes[0].axhline(0, color='gray', linestyle='--')
axes[0].set_title("Residuos vs ajustados (homoscedasticidad)")
axes[0].set_xlabel("Valores ajustados")
axes[0].set_ylabel("Residuos")

# Heterocedasticidad
sns.scatterplot(x=model_hetero.fittedvalues, y=model_hetero.resid, ax=axes[1], color='darkorange')
axes[1].axhline(0, color='gray', linestyle='--')
axes[1].set_title("Residuos vs ajustados (heterocedasticidad)")
axes[1].set_xlabel("Valores ajustados")
axes[1].set_ylabel("Residuos")

plt.tight_layout()
plt.show()

#  Prueba de Breusch‚ÄìPagan
bp_homo = het_breuschpagan(model_homo.resid, model_homo.model.exog)
bp_hetero = het_breuschpagan(model_hetero.resid, model_hetero.model.exog)

print(" Breusch‚ÄìPagan (homoscedasticidad):")
print(f"p-valor: {bp_homo[1]:.4f} ‚Üí {'No hay evidencia de heterocedasticidad' if bp_homo[1] > 0.05 else 'Posible heterocedasticidad'}")

print("\n Breusch‚ÄìPagan (heterocedasticidad):")
print(f"p-valor: {bp_hetero[1]:.4f} ‚Üí {'No hay evidencia de heterocedasticidad' if bp_hetero[1] > 0.05 else 'Posible heterocedasticidad'}")
```

<div style="background-color:#f2f2f2; padding:10px; border-radius:5px;">
  <b>Interpretaci√≥n del supuesto de homoscedasticidad:</b><br>
  <ul>
    <li>En el caso homosced√°stico, los residuos se distribuyen de forma uniforme a lo largo de los valores ajustados. Esto indica que la varianza de los errores es constante y el supuesto se cumple.</li>
    <li>En el caso heterosced√°stico, los residuos se expanden en forma de abanico o embudo. Esto sugiere que la varianza de los errores aumenta con el valor de la variable predictora, lo que viola el supuesto.</li>
    <li>La prueba de Breusch‚ÄìPagan permite contrastar esta diferencia. Un p-valor alto indica que no hay evidencia de heterocedasticidad, mientras que un p-valor bajo indica que la varianza de los errores depende de los predictores.</li>
    <li>Cuando se detecta heterocedasticidad, se recomienda ajustar el modelo utilizando errores robustos o aplicar transformaciones a la variable dependiente para estabilizar la varianza.</li>
  </ul>
</div>



4. **Normalidad:** los errores deben seguir una distribuci√≥n normal (relevante para intervalos de confianza e inferencia) [(Shapiro & Wilk, 1965)](https://doi.org/10.1093/biomet/52.3-4.591).  T√©cnicas de evaluaci√≥n:
     - **Histograma de residuos:** debe asemejarse a una curva normal (en forma de campana).  
     - **Gr√°fico Q‚ÄìQ (Quantile-Quantile):** compara cuantiles de los residuos con cuantiles de una distribuci√≥n normal; puntos cercanos a la l√≠nea diagonal indican normalidad.  
     - **Prueba de Shapiro‚ÄìWilk:** estad√≠stico espec√≠fico para contrastar normalidad (p-valor > 0.05 ‚Üí se asume normalidad).  
     - **Prueba de Kolmogorov‚ÄìSmirnov:** alternativa para grandes muestras, aunque menos potente que Shapiro‚ÄìWilk.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import shapiro
from statsmodels.graphics.gofplots import qqplot

np.random.seed(42)
n = 200
X = np.random.normal(50, 10, n)

# Caso con errores normales
error_normal = np.random.normal(0, 5, n)
Y = 3 + 0.7 * X + error_normal

df = pd.DataFrame({"X": X, "Y": Y})

X_const = sm.add_constant(df["X"])
model = sm.OLS(df["Y"], X_const).fit()
resid = model.resid

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histograma
sns.histplot(resid, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title("Histograma de residuos")
axes[0].set_xlabel("Residuos")
axes[0].set_ylabel("Frecuencia")

# Q‚ÄìQ plot
qqplot(resid, line='s', ax=axes[1])
axes[1].set_title("Gr√°fico Q‚ÄìQ de residuos")

plt.tight_layout()
plt.show()

# Prueba de Shapiro‚ÄìWilk
stat, p_value = shapiro(resid)
print(f"Prueba de Shapiro‚ÄìWilk")
print(f"Estad√≠stico: {stat:.4f}")
print(f"p-valor: {p_value:.4f} ‚Üí {'Se asume normalidad' if p_value > 0.05 else 'No se cumple el supuesto de normalidad'}")
```

<div style="background-color:#f2f2f2; padding:10px; border-radius:5px;">
  <b>Interpretaci√≥n del supuesto de normalidad:</b><br>
  <ul>
    <li>En el histograma de residuos, se espera una forma de campana sim√©trica. Si hay asimetr√≠a o picos extremos, puede indicar desviaci√≥n de la normalidad.</li>
    <li>En el gr√°fico Q‚ÄìQ, los puntos deben alinearse con la diagonal. Desviaciones sistem√°ticas (curvas o extremos alejados) sugieren que los residuos no siguen una distribuci√≥n normal.</li>
    <li>La prueba de Shapiro‚ÄìWilk permite contrastar estad√≠sticamente la normalidad. Un p-valor mayor a 0.05 indica que no hay evidencia suficiente para rechazar la normalidad.</li>
    <li>La prueba de Kolmogorov‚ÄìSmirnov es √∫til en muestras grandes, aunque menos sensible a desviaciones sutiles que Shapiro‚ÄìWilk.</li>
  </ul>
</div>


5. **Ausencia de multicolinealidad:** los predictores no deben estar altamente correlacionados entre s√≠  [(O‚ÄôBrien, 2007)](https://doi.org/10.3102/1076998606298031). T√©cnicas de evaluaci√≥n:
     - **Matriz de correlaci√≥n:** √∫til como diagn√≥stico inicial; correlaciones altas (>0.8) sugieren riesgo de multicolinealidad.  
     - **VIF (Variance Inflation Factor):** mide cu√°nto aumenta la varianza de un coeficiente por colinealidad; valores VIF > 10 indican problema serio.  
     - **Condici√≥n del n√∫mero de la matriz (Condition Index):** valores altos (>30) se√±alan fuerte dependencia lineal entre predictores.  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from numpy.linalg import svd

np.random.seed(42)
n = 200

# Caso 1: sin multicolinealidad
X1 = np.random.normal(50, 10, n)
X2 = np.random.normal(30, 5, n)  # independiente de X1
Y1 = 5 + 0.6 * X1 + 0.4 * X2 + np.random.normal(0, 5, n)

# Caso 2: con multicolinealidad
X3 = np.random.normal(50, 10, n)
X4 = X3 + np.random.normal(0, 2, n)  # altamente correlacionado con X3
Y2 = 5 + 0.6 * X3 + 0.4 * X4 + np.random.normal(0, 5, n)

df_sin = pd.DataFrame({"X1": X1, "X2": X2, "Y": Y1})
df_con = pd.DataFrame({"X3": X3, "X4": X4, "Y": Y2})

# Matriz de correlaci√≥n
print("Correlaci√≥n entre predictores (sin multicolinealidad):")
print(df_sin[["X1", "X2"]].corr(), "\n")

print("Correlaci√≥n entre predictores (con multicolinealidad):")
print(df_con[["X3", "X4"]].corr(), "\n")

# C√°lculo de VIF (Variance Inflation Factor)
def calcular_vif(X):
    X_const = sm.add_constant(X)
    vif = pd.DataFrame()
    vif["Variable"] = X_const.columns
    vif["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
    return vif

print("VIF (sin multicolinealidad):")
print(calcular_vif(df_sin[["X1", "X2"]]), "\n")

print("VIF (con multicolinealidad):")
print(calcular_vif(df_con[["X3", "X4"]]), "\n")

# C√°lculo del √≠ndice de condici√≥n
def calcular_indice_condicion(X):
    X_const = sm.add_constant(X)
    _, s, _ = svd(X_const)
    return max(s) / min(s)

print("√çndice de condici√≥n (sin multicolinealidad):")
print(f"{calcular_indice_condicion(df_sin[['X1', 'X2']]):.2f}\n")

print("√çndice de condici√≥n (con multicolinealidad):")
print(f"{calcular_indice_condicion(df_con[['X3', 'X4']]):.2f}\n")
```

<div style="background-color:#f2f2f2; padding:10px; border-radius:5px;">
  <b>Interpretaci√≥n del supuesto de ausencia de multicolinealidad:</b><br>
  <ul>
    <li>En el primer caso, los predictores X1 y X2 son independientes. La matriz de correlaci√≥n muestra valores bajos, los VIF est√°n cerca de 1 y el √≠ndice de condici√≥n es bajo. Esto indica que no hay multicolinealidad.</li>
    <li>En el segundo caso, X3 y X4 est√°n altamente correlacionados. La matriz de correlaci√≥n muestra valores cercanos a 1, los VIF son elevados y el √≠ndice de condici√≥n supera el umbral cr√≠tico. Esto indica multicolinealidad severa.</li>
  </ul>
</div>

## <span style="color:black;"><strong>3.3. An√°lisis de cl√∫ster</strong></span>  

El **an√°lisis de cl√∫ster** es un m√©todo de clasificaci√≥n no supervisada que busca agrupar observaciones en **cl√∫steres homog√©neos**, de manera que los objetos dentro de un mismo grupo sean lo m√°s parecidos posible, y lo m√°s diferentes entre grupos [(Hair et al., 2019)](https://doi.org/10.1007/978-3-030-06031-2).  

Una medida com√∫n para cuantificar la similitud entre observaciones es la **distancia euclidiana** entre dos individuos $i$ y $j$:  

$$
d(i,j) = \sqrt{\sum_{k=1}^{p}(x_{ik} - x_{jk})^2}
$$  

donde $p$ representa el n√∫mero de variables y $x_{ik}$ el valor de la $k$-√©sima variable para la observaci√≥n $i$.  


### <span style="color:#2F749F;"><strong>Fundamentos matem√°ticos</strong></span>  

El an√°lisis de cl√∫ster se basa en la definici√≥n de una **medida de distancia o similitud** y en un **criterio de agrupamiento** [(Everitt et al., 2011)](https://doi.org/10.1002/9780470977811).  

1. **Matriz de distancias:** a partir de los datos estandarizados, se calcula una matriz $D$ de distancias $d(i,j)$ entre cada par de observaciones.  
2. **Criterio de agrupamiento:** se selecciona un algoritmo que defina c√≥mo se forman los grupos:  
   - **Jer√°rquico (aglomerativo o divisivo):** construye un dendrograma que refleja el proceso de uni√≥n o divisi√≥n de cl√∫steres.  
   - **K-means:** minimiza la suma de distancias al centroide de cada cl√∫ster. Su objetivo es:  

   $$
   \min \sum_{k=1}^{K} \sum_{x_i \in C_k} ||x_i - \mu_k||^2
   $$  

   donde $C_k$ es el cl√∫ster $k$ y $\mu_k$ su centroide.  
   - **DBSCAN:** identifica regiones densas de observaciones y permite detectar ruido o puntos at√≠picos [(Ester et al., 1996)](https://www.aaai.org/Papers/KDD/1996/KDD96-037.pdf).  

3. **N√∫mero de cl√∫steres:** se determina mediante criterios como el **m√©todo del codo** o el **√≠ndice de silueta**, que eval√∫an la coherencia interna de los grupos.  



### <span style="color:#2F749F;"><strong>Interpretaci√≥n de los cl√∫steres</strong></span>  

- **Centroides ($\mu_k$):** representan el punto promedio de cada cl√∫ster (en m√©todos como K-means).  
- **Tama√±o de cl√∫ster:** indica cu√°ntas observaciones pertenecen a cada grupo.  
- **√çndice de silueta:** mide la calidad de la agrupaci√≥n. Para una observaci√≥n $i$:  

$$
s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
$$  

donde:  
- $a(i)$ es la distancia promedio de $i$ a los puntos de su mismo cl√∫ster.  
- $b(i)$ es la distancia promedio de $i$ al cl√∫ster m√°s cercano distinto.  

El valor de $s(i)$ var√≠a entre -1 y 1, donde valores cercanos a 1 indican una asignaci√≥n correcta y valores negativos sugieren una mala clasificaci√≥n [(Rousseeuw, 1987)](https://doi.org/10.1016/0377-0427(87)90125-7).  


```python
# Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist
from sklearn.preprocessing import MinMaxScaler

datos = {
    'Nombre': ['Alan', 'Lisa', 'Joe', 'Max', 'Cora', 'Adam', 'Kim', 'Ali', 'Chen', 'Jack', 'Levi'],
    'Social_Media': [7, 5, 5, 7, 4, 8, 2, 8, 4, 14, 8],
    'Gimnasio': [3, 2, 3, 4, 5, 9, 12, 2, 14, 1, 10]
}
df = pd.DataFrame(datos)

print(df.head())

# Gr√°fico exploratorio
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='Social_Media', y='Gimnasio', color='steelblue')
for i in range(len(df)):
    plt.text(df['Social_Media'][i]+0.2, df['Gimnasio'][i], df['Nombre'][i], fontsize=9)
plt.title('Distribuci√≥n inicial')
plt.xlabel('Horas en redes sociales')
plt.ylabel('Horas en gimnasio')
plt.tight_layout()
plt.show()


```

<div style="background-color:#fff9db; padding:20px; border-radius:10px;">
<b>Inicio del algoritmo:</b>  

1. Si se cuenta con 11 personas, el procedimiento comienza con 11 conglomerados individuales.  
2. En cada iteraci√≥n, el algoritmo fusiona los dos conglomerados m√°s pr√≥ximos seg√∫n la m√©trica de distancia seleccionada (por ejemplo, euclidiana).  
3. Este proceso se repite de manera sucesiva hasta que:  
   - Todas las observaciones pertenecen a un √∫nico conglomerado, o  
   - Se define un punto de corte en el dendrograma que determina el n√∫mero de grupos finales.  
</div>


```python
print("N√∫mero inicial de conglomerados:", len(df))
```

```python
# Estandarizar 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Social_Media', 'Gimnasio']])

# Matriz de distancias
dist_matrix = pdist(X_scaled, metric='euclidean')

# Enlace jer√°rquico
linked = linkage(dist_matrix, method='ward')
# Normalizar las distancias del dendrograma
linked_scaled = linked.copy()
linked_scaled[:, 2] = MinMaxScaler().fit_transform(linked[:, 2].reshape(-1, 1)).flatten()

# Definir umbral entre 0 y 1
umbral = 0.6

# Visualizar dendrograma
plt.figure(figsize=(10, 5))
dendrogram(linked_scaled, labels=df['Nombre'].values, leaf_rotation=90, color_threshold=umbral)
plt.axhline(y=umbral, color='gray', linestyle='--', label=f'Umbral = {umbral}')
plt.title('Dendrograma')
plt.xlabel('Nombre')
plt.ylabel('Distancia')
plt.legend()
plt.tight_layout()
plt.show()

# Asignar grupos seg√∫n el umbral
df['Grupo_Jerarquico'] = fcluster(linked_scaled, t=umbral, criterion='distance')

# Imprimir resumen de grupos
print(df[['Nombre', 'Social_Media', 'Gimnasio', 'Grupo_Jerarquico']].sort_values('Grupo_Jerarquico'))
```

## <span style="color:black;"><strong>3.4. An√°lisis factorial</strong></span>  

El **an√°lisis factorial** es una t√©cnica multivariada que busca identificar **factores latentes** que explican las correlaciones entre un conjunto de variables observadas. Su prop√≥sito principal es reducir la complejidad de los datos, describi√©ndolos en t√©rminos de un menor n√∫mero de dimensiones subyacentes [(Fabrigar & Wegener, 2011)](https://doi.org/10.1093/acprof:osobl/9780199734177.001.0001).  

El modelo factorial se expresa como:  

$$
X = \Lambda F + \varepsilon
$$  

donde:  
- $X \in \mathbb{R}^p$: vector de variables observadas.  
- $F \in \mathbb{R}^m$: vector de factores comunes ($m < p$).  
- $\Lambda \in \mathbb{R}^{p \times m}$: matriz de cargas factoriales.  
- $\varepsilon \in \mathbb{R}^p$: vector de errores espec√≠ficos.  


### <span style="color:#2F749F;"><strong>Fundamentos matem√°ticos</strong></span>  

El an√°lisis factorial se basa en la descomposici√≥n de la **matriz de covarianza o correlaci√≥n** de los datos en varianza com√∫n y varianza espec√≠fica [(Anderson, 2003)](https://doi.org/10.1002/0471264028).  

1. **Supuestos b√°sicos:**  
   - $\mathbb{E}[F] = 0$, $\text{Cov}(F) = I_m$ (factores estandarizados e independientes).  
   - $\mathbb{E}[\varepsilon] = 0$, $\text{Cov}(\varepsilon) = \Psi$ (matriz diagonal de varianzas espec√≠ficas).  
   - $\text{Cov}(F,\varepsilon) = 0$ (independencia entre factores y errores).  

2. **Descomposici√≥n de la varianza:**  
   Bajo estos supuestos, la matriz de covarianza $\Sigma$ de $X$ se puede escribir como:  

   $$
   \Sigma = \Lambda \Lambda^\top + \Psi
   $$  

   donde $\Lambda \Lambda^\top$ representa la varianza com√∫n explicada por los factores y $\Psi$ la varianza √∫nica de cada variable.  

3. **M√©todos de estimaci√≥n:**  
   - **M√°xima verosimilitud (ML):** optimiza la funci√≥n de verosimilitud bajo normalidad multivariante.  
   - **M√≠nimos cuadrados:** minimiza la diferencia entre la matriz observada y la estimada.  
   - **Componentes principales (ACP):** aproximaci√≥n inicial para extraer factores.  

4. **Rotaci√≥n de factores:** t√©cnicas como **Varimax** (ortogonal) u **Oblimin** (oblicua) permiten simplificar la estructura de cargas factoriales y facilitar la interpretaci√≥n.  


### <span style="color:#2F749F;"><strong>Interpretaci√≥n de los factores</strong></span>  

- **Cargas factoriales ($\lambda_{jm}$):** representan la correlaci√≥n entre la variable $j$ y el factor $m$.  
- **Comunalidad ($h_j^2$):** proporci√≥n de la varianza de la variable $j$ explicada por los factores:  

  $$
  h_j^2 = \sum_{m=1}^{M} \lambda_{jm}^2
  $$  

- **Varianza √∫nica ($\psi_j$):** parte de la varianza de $X_j$ no explicada por los factores.  
- **N√∫mero de factores:** se determina mediante criterios como el **autovalor > 1** (criterio de Kaiser) o el **gr√°fico de sedimentaci√≥n**.  

El an√°lisis factorial es ampliamente usado en ciencias sociales, psicometr√≠a y estudios de mercado para identificar dimensiones ocultas y construir escalas de medici√≥n v√°lidas [(Hair et al., 2019)](https://doi.org/10.1007/978-3-030-06031-2).  


```python
# Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer

datos = {
    'Nombre': ['Alan', 'Lisa', 'Joe', 'Max', 'Cora', 'Adam', 'Kim', 'Ali', 'Chen', 'Jack', 'Levi'],
    'Social_Media': [7, 5, 5, 7, 4, 8, 2, 8, 4, 14, 8],
    'Gimnasio': [3, 2, 3, 4, 5, 9, 12, 2, 14, 1, 10]
}
df = pd.DataFrame(datos)

# Estandarizar variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Social_Media', 'Gimnasio']])
X_df = pd.DataFrame(X_scaled, columns=['Social_Media', 'Gimnasio'])

# Matriz de correlaciones
print("\n Matriz de correlaciones:")
print(X_df.corr())

# Aplicar an√°lisis factorial
fa = FactorAnalyzer(n_factors=1, rotation=None)
fa.fit(X_scaled)

# Cargas factoriales
cargas = fa.loadings_
print("\n Cargas factoriales:")
print(pd.DataFrame(cargas, index=['Social_Media', 'Gimnasio'], columns=['Factor 1']))

# Varianza explicada
var_exp = fa.get_factor_variance()
print("\n Varianza explicada:")
print(pd.DataFrame({
    'Varianza': var_exp[0],
    'Proporci√≥n': var_exp[1],
    'Acumulada': var_exp[2]
}, index=['Factor 1']))

# Visualizaci√≥n de cargas factoriales
plt.figure(figsize=(5, 4))
ax = sns.barplot(x=['Social_Media', 'Gimnasio'], y=cargas.flatten(), legend=False)
for i, val in enumerate(cargas.flatten()):
    ax.text(i, val + 0.05*np.sign(val), f'{val:.2f}', ha='center', va='bottom', fontsize=10)
plt.title('Cargas factoriales sobre Factor 1')
plt.ylabel('Carga')
plt.ylim(-1, 1)
plt.tight_layout()
plt.show()
```
